# NBC News Scraping

This notebook runs the NBC scraper script and inspects the saved dataset.

from pathlib import Path
import subprocess
import sys

repo_root = Path.cwd().resolve().parent
script_path = repo_root / "scripts" / "scrape_nbc.py"
output_path = repo_root / "data" / "raw" / "nbc_scraped_all.csv"

print(f"Running: {script_path}")
result = subprocess.run([sys.executable, str(script_path)], cwd=repo_root, check=False)
print(f"Exit code: {result.returncode}")
print(f"Output file expected at: {output_path}")

```python
import pandas as pd

df = pd.read_csv(output_path)
print(f"Rows: {len(df)}")
print(f"Columns: {list(df.columns)}")

df.head()
```
