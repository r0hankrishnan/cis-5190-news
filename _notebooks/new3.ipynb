```python
from sentence_transformers import SentenceTransformer
import numpy as np
import pandas as pd

feature_df = pd.read_csv("../data/processed/style_features.csv", index_col=0)

embedder = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = embedder.encode(feature_df["title"].fillna("").tolist(), show_progress_bar=True)

np.save("../data/processed/minilm_embeddings.npy", embeddings)
print(f"Saved embeddings shape: {embeddings.shape}")
```
